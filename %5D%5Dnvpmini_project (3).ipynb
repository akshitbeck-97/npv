##### Step1: Display the suitable graph for the number of calories in Crispy Chicken category

```python
import matplotlib.pyplot as plt
plt.bar(d1.groupby('Category_Type')['Calories'].min().index,d1.groupby('Category_Type')['Calories'].min().values)
plt.xticks(rotation=45)
plt.show()

```

```python
d1
```

##### Step2: Find the average calories for Crispy Chicken Category

```python
d1[d1['Item_in_Menu'].str.find('Crispy Chicken')!=-1]['Calories'].mean()
```


 Step3: Display the suitable graph for the number of calories in Grilled Chicken category

```python
import matplotlib.pyplot as plt
plt.bar(d1[d1['Item_in_Menu'].str.find('Crispy Chicken')!=-1]['Item_in_Menu'],d1[d1['Item_in_Menu'].str.find('Crispy Chicken')!=-1]['Calories'])
plt.xticks(rotation=45)
plt.show()
```

Step4:Find the average calories for Grilled Chicken Category

```python
d1[d1['Item_in_Menu'].str.find('Grilled Chicken')!=-1]['Calories'].mean()
```

##### Step 5: Get the final visual comparison of Crispy vs Grilled on the basis of calories

```python
d1['Calories'].values
```

```python
plt.plot(d1[d1['Item_in_Menu'].str.find('Grilled Chicken')!=-1]['Calories'],label='Grilled chicken')
plt.plot(d1[d1['Item_in_Menu'].str.find('Crispy Chicken')!=-1]['Calories'],label='Crispy chicken')
plt.legend()
plt.xlabel('no of items')
plt.ylabel('calories')
plt.show()
```
